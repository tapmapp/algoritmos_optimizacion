```
from pprint import pprint
import numpy as np
import random

poblacionNum = 500
poblacion = []
machos = []
hembras = []

# Variables para la generacion de pesos aleatorios
decimales = 2
mediaPeso = 37
desvPeso = 2.1

# Generamos los pesos de la poblacion en funcion de los valores del enunciado
peso = np.random.normal(mediaPeso, desvPeso, poblacionNum)

# Generar poblacion aleatoria
def generarPoblacion(poblacionNum):

    for n in range(len(peso)):

        # generamos el sexo de manera aleatoria 0 para hembras, 1 para machos
        randomNum = random.randint(0, 1)

        if(randomNum == 0):
            sexo = 'H'
        else:
            sexo = 'M'

        # obtenmos la desviacion de cada elemento
        desviacion = abs(peso[n] - peso.mean())

        # anadimos los valores a la lista poblacion
        poblacion.append({ 'id': n, 'peso': round(peso[n], 1), 'sexo': sexo, 'desviacion': round(desviacion, 2) })
    
    # ordenamos los valores en funcion de su diferencia con respecto a la media de menor a mayor
    # para la posterior generacion de grupos y eliminacion de valores
    poblacion.sort(key=lambda x: x['desviacion'])

    # clasificamos los elementos 
    clasificarElementos(poblacion)

# clasificador de la poblacion
def clasificarElementos(poblacion):

    npPoblacion = np.array([d['peso'] for d in poblacion])

    pesoMediaPoblacion = npPoblacion.mean()
    pesoVarianzaPoblacion = npPoblacion.var()
    pesoDesviacionPoblacion = npPoblacion.std()

    print('')
    print('POBLACION TOTAL')
    print('Peso Medio: ' + str(round(pesoMediaPoblacion, 2)))
    print('Varianza: ' + str(round(pesoVarianzaPoblacion, 2)))
    print('Desviacion: ' + str(round(pesoDesviacionPoblacion, 2)))
    
    for n in range(len(poblacion)):
        
        # clasificamos la poblacion en funcion del sexo y eliminamos los valores que superan la desviacion
        if(poblacion[n]['sexo'] == 'H'):
            hembras.append(poblacion[n])

        if(poblacion[n]['sexo'] == 'M'):
            machos.append(poblacion[n])

    # obtenemos cual es la lista con menos elementos para posteriormente crear grupos con los mismos elementos
    if(len(hembras) >= len(machos)):
        
        # obtenemos un numero que sea divisble entre 9 para generar los grupos
        if(len(machos) % 9 == 0):
            tamanoValido = len(machos)
        else:

            i = len(machos)
            while i % 9 != 0:
                i -= 1

            tamanoValido = i

    else:

        # obtenemos un numero que sea divisble entre 9 para generar los grupos
        if(len(hembras) % 9 == 0):
            tamanoValido = len(hembras)
        else:

            i = len(hembras)
            while i % 9 != 0:
                i -= 1

            tamanoValido = i

    # eliminamos los valores con mas diferencia a la media utlizando el tamano de lista calculado previamente
    # y cambiamos el orden de los elementos de forma aleatoria
    hembrasTrim = hembras[:tamanoValido]
    machosTrim = machos[:tamanoValido]

    print('')
    print('HEMBRAS')
    generarGrupos(hembrasTrim)

    print('')
    print('MACHOS')
    generarGrupos(machosTrim)

# generador de grupos
def generarGrupos(poblacion):

    # calculamos el numero de elementos para 9 grupos
    elementosPorGrupo = len(poblacion)//9

    grupos = []
    grupo = []
    
    # organizamos la poblacion en grupos de 9
    for n in range(len(poblacion)):
        grupo.append(poblacion[n])
        if(len(grupo) == elementosPorGrupo):
            grupos.append(grupo)
            grupo = []

    clasificadorGrupos(grupos, elementosPorGrupo)

# clasificador grupos
def clasificadorGrupos(grupos, elementosPorGrupo):

    nuevosGrupos = []
    grupo = []

    # organizamos la poblacion en grupos homogeneos
    # recorremos la lista grupos el numero de veces que elementos tenga por grupo
    for i in range(elementosPorGrupo):
        for n in range(len(grupos)):
            for m in range(len(grupos[n])):
                if(i == m and len(grupo) < elementosPorGrupo):
                    grupo.append(grupos[n][m])
                if(len(grupo) == elementosPorGrupo):
                    nuevosGrupos.append(grupo)
                    grupo = []

    clasificadorTratamientos(nuevosGrupos)

# clasificador tratamientos
def clasificadorTratamientos(grupos):

    tratamientos = []
    grupo = []

    # organizamos los grupos por tratamiento
    for n in range(len(grupos)):
        grupo.append(grupos[n])
        if(len(grupo) == 3):
            tratamientos.append(grupo)
            grupo = []
    
    # mostramos los resultados
    for n in range(len(tratamientos)):
        mostrarResultados(tratamientos[n], n)

        
# mostrar resultados
def mostrarResultados(tratamiento, indice):

    print('')
    print('TRATAMIENTO ' + str(indice + 1))

    for n in range(len(tratamiento)):

        npTratamiento = np.array([d['peso'] for d in tratamiento[n]])

        pesoMedioPoblacion = npTratamiento.mean()
        pesoVarianzaPoblacion = npTratamiento.var()
        pesoDesviacionPoblacion = npTratamiento.std()

        print('')
        print('GRUPO: ' + str(n + 1))
        print('Peso Medio: ' + str(round(pesoMedioPoblacion, 2)))
        print('Varianza: ' + str(round(pesoVarianzaPoblacion, 2)))
        print('Desviacion: ' + str(round(pesoDesviacionPoblacion, 2)))

# iniciamos el algoritmo
generarPoblacion(poblacionNum)
```

###**Diseña un algoritmo que mejore la complejidad del algoritmo por fuerza bruta. Argumenta por qué crees que mejora el algoritmo por fuerza bruta.**

Considero que el algoritmo mejora  el algoritmo de fuerza bruta porque hay menos bucles por lo que la complejidad disminuye y aumenta en eficiencia y rendimiento.

```
from pprint import pprint
import numpy as np
import random

poblacionNum = 1000
poblacion = []
machos = []
hembras = []

# Variables para la generacion de pesos aleatorios
decimales = 2
mediaPeso = 37
desvPeso = 2.1

# Generamos los pesos de la poblacion en funcion de los valores del enunciado
peso = np.random.normal(mediaPeso, desvPeso, poblacionNum)

# Generar poblacion aleatoria
def generarPoblacion(poblacionNum):

    for n in range(len(peso)):

        # generamos el sexo de manera aleatoria 0 para hembras, 1 para machos
        randomNum = random.randint(0, 1)

        if(randomNum == 0):
            sexo = 'H'
        else:
            sexo = 'M'

        # obtenmos la desviacion de cada elemento
        desviacion = abs(peso[n] - peso.mean())

        # anadimos los valores a la lista poblacion
        poblacion.append({ 'id': n, 'peso': round(peso[n], 1), 'sexo': sexo, 'desviacion': round(desviacion, 2) })
    
    # ordenamos los valores en funcion de su diferencia con respecto a la media de menor a mayor
    # para la posterior generacion de grupos y eliminacion de valores
    poblacion.sort(key=lambda x: x['desviacion'])

    # clasificamos los elementos 
    clasificarElementos(poblacion)

# clasificador de la poblacion
def clasificarElementos(poblacion):

    npPoblacion = np.array([d['peso'] for d in poblacion])

    pesoMediaPoblacion = npPoblacion.mean()
    pesoVarianzaPoblacion = npPoblacion.var()
    pesoDesviacionPoblacion = npPoblacion.std()

    print('')
    print('POBLACION TOTAL: ' + str(len(npPoblacion)))
    print('Peso Medio: ' + str(round(pesoMediaPoblacion, 2)))
    print('Varianza: ' + str(round(pesoVarianzaPoblacion, 2)))
    print('Desviacion: ' + str(round(pesoDesviacionPoblacion, 2)))
    
    for n in range(len(poblacion)):
        
        # clasificamos la poblacion en funcion del sexo y eliminamos los valores que superan la desviacion
        if(poblacion[n]['sexo'] == 'H'):
            hembras.append(poblacion[n])

        if(poblacion[n]['sexo'] == 'M'):
            machos.append(poblacion[n])

    # obtenemos cual es la lista con menos elementos para posteriormente crear grupos con los mismos elementos
    if(len(hembras) >= len(machos)):
        
        # obtenemos un numero que sea divisble entre 9 para generar los grupos
        if(len(machos) % 9 == 0):
            tamanoValido = len(machos)
        else:

            i = len(machos)
            while i % 9 != 0:
                i -= 1

            tamanoValido = i

    else:

        # obtenemos un numero que sea divisble entre 9 para generar los grupos
        if(len(hembras) % 9 == 0):
            tamanoValido = len(hembras)
        else:

            i = len(hembras)
            while i % 9 != 0:
                i -= 1

            tamanoValido = i

    hembrasTrim = hembras[:tamanoValido]
    machosTrim = machos[:tamanoValido]

    print('')
    print('HEMBRAS')
    generarGrupos(hembrasTrim)

    print('')
    print('MACHOS')
    generarGrupos(machosTrim)

# generador de grupos
def generarGrupos(poblacion):

    # calculamos el numero de elementos para 9 grupos
    elementosPorGrupo = len(poblacion)//9
    
    # dividimos la poblacion en grupos
    # los grupos estaran compuestos por elementos de similar desviacion
    grupos = dividirGrupos([], 9, elementosPorGrupo, poblacion)

    # organizamos la poblacion en grupos homogeneos y uniformes
    # tratando que todos los grupos tengan elementos similares
    clasificadorGrupos(grupos, elementosPorGrupo)

# divisor de elementos por grupo en funcion de su desviacion
def dividirGrupos(grupos, numeroGrupos, elementosPorGrupo, poblacion):
    
    # separamos la poblacion
    leftPart = poblacion[:elementosPorGrupo]
    rightPart = poblacion[elementosPorGrupo:]

    # cambiamos el orden de los elementos de manera aleatoria
    random.shuffle(leftPart)

    # anadimos el primer bloque 
    grupos.append(leftPart)

    # aplicamos recursividad mientras el numero de elementos sobrantes sea mayor
    # que el numero de elementos por grupo
    if(len(grupos) < numeroGrupos):
        dividirGrupos(grupos, numeroGrupos, elementosPorGrupo, rightPart)

    return grupos

# clasificador grupos
def clasificadorGrupos(grupos, elementosPorGrupo):

    # generamos contadores para acceder a los valores de los grupos
    valueCounter = 0 
    groupCounter = 0

    # generamos los grupos homogeneos
    nuevosGrupos = readGroup([], [], grupos, valueCounter, groupCounter, elementosPorGrupo)

    # clasificamos los tratamientos
    clasificadorTratamientos(nuevosGrupos)

def readGroup(nuevosGrupos, grupo, grupos, valueCounter, groupCounter, elementosPorGrupo):

    # vamos recorriendo cada grupo y anadiendo un valor por grupo a una lista grupo
    for n in range(len(grupos)):
        grupo.append(grupos[n][valueCounter])
    
    # cuanda la lista alcanza el numero de elementos por grupo 
    # anadimos la lista a otra lista llamada nuevos grupos donde guardaremos todos los grupos
    # y vaciamos la lista  
    if(len(grupo) == elementosPorGrupo):
        nuevosGrupos.append(grupo)
        grupo = []

    # aplicamos recursividad mientras el contador sea menor que el numero de elementos por grupo
    if(valueCounter < elementosPorGrupo - 1):
        
        # vamos anadiendo valores al contador para ir accediendo a los valores de los grupos
        valueCounter += 1
        
        # aplicamos recursividad
        readGroup(nuevosGrupos, grupo, grupos, valueCounter, groupCounter, elementosPorGrupo)        

    return nuevosGrupos

# clasificador tratamientos
def clasificadorTratamientos(grupos):

    # generamos los grupos para los tratamientos aplicando recursividad con el metodo anterior
    tratamientos = dividirGrupos([], 3, 3, grupos)

    # mostramos los resultados
    for n in range(len(tratamientos)):
        mostrarResultados(tratamientos[n], n)
        
# mostrar resultados
def mostrarResultados(tratamiento, indice):

    print('')
    print('TRATAMIENTO ' + str(indice + 1))

    for n in range(len(tratamiento)):

        npTratamiento = np.array([d['peso'] for d in tratamiento[n]])

        pesoMedioPoblacion = npTratamiento.mean()
        pesoVarianzaPoblacion = npTratamiento.var()
        pesoDesviacionPoblacion = npTratamiento.std()

        print('')
        print('GRUPO: ' + str(n + 1))
        print('Elementos por grupo: ' +  str(len(tratamiento[n])))
        print('Peso Medio: ' + str(round(pesoMedioPoblacion, 2)))
        print('Varianza: ' + str(round(pesoVarianzaPoblacion, 2)))
        print('Desviacion: ' + str(round(pesoDesviacionPoblacion, 2)))

# iniciamos el algoritmo
generarPoblacion(poblacionNum)
```
